```python
import math
import random
import time
```

# Algoritmo de Fermat

```python
def fermat(n):
  x = math.ceil(math.sqrt(n))
  y_2 = x**2 - n

  def is_perfect_square(z):
    return math.isqrt(z)**2 == z

  while(not is_perfect_square(y_2)):
    x += 1
    y_2 = x**2 - n

  y = math.sqrt(y_2)
  p = x + y
  q = x - y

  print(f"p = {p}, q = {q}")
```

# Reducción de factorización a raíz cuadrada

```python
def reduccion(n):
  y = random.randint(2, n-2)
  a = (y**2)%n
  possible_x = []

  for x in range(0,n):
    if ((x**2)%n)%a == 0: possible_x.append(x)

  for x_p in possible_x:
    if x_p == 0:
      continue
    if ((y%n)%x_p != 0 and ((-y)%n)%x_p != 0):
      p = math.gcd(x_p-y,n)
      q = n/p
      print(f"p = {p}, q = {q}")
      return

```

# Pruebas con diferentes valores de n

```python
def benchmark(arr):
  for value in arr:
    print("N: ",value)

    print("Solución mediante Fermat: ")
    start_fermat = time.time()
    fermat(value)
    end_fermat = time.time()
    print(f"Tiempo de ejecución de Fermat: {end_fermat - start_fermat:.6f} segundos\n")

    print("Solución mediante Reducción: ")
    start_reduccion = time.time()
    reduccion(value)
    end_reduccion = time.time()
    print(f"Tiempo de ejecución de Reducción: {end_reduccion - start_reduccion:.6f} segundos\n")

    print("-----------------------------------------\n")
```

```python
n_values = [21, 22275377, 72401183, 8904031, 95532967, 40053991, 165725927, 173640199, 221777863, 299526169]
```

```python
benchmark(n_values)
```
